```python
# cleaning of data
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the mode
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' column due to too many missing values
# Drop 'Cabin' column due to too many missing values
data.drop('Cabin', axis=1, inplace=True)
```

```python
# checking if any null values are still remaining in the data set
print(data.isnull().sum())
```

```python
data.hist(bins=10, figsize=(10,6))
```

```python
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Count by Sex')
plt.show()
```

```python
# EDA for survival count by age ( understanding the survival pattern with the help of age variable)
sns.histplot(x='Age', hue='Survived', data=data, kde=True)
plt.title('Survival Count by Age')
plt.show()
```

```python
# Visualizing survival rate by embarkation port
sns.countplot(x='Survived', hue='Embarked', data=data)
plt.title('Survival Count by Embarkation Port')
plt.show()
```

```python
survival_rate_by_class = data.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values)
plt.title('Survival Rate by Pclass')
```
